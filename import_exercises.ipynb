Read about and use the itertools module from the python standard library to help you solve the following problems (Hint: wrap the itertools results in a list to see the results).



- How many different ways can you combine a single letter from "abc" with either 1, 2, or 3?


```python
import itertools
letters = 'abc'
numbers = [1,2,3]

combinations = list(itertools.product(letters, numbers))

print(len(combinations))
print(combinations)
```

- How many different combinations are there of 2 letters from "abcd"?


```python
import itertools
letters2 = 'abcd'

combinations2 = list(itertools.combinations(letters2, 2))

print(combinations2)
print(len(combinations2))
```

- How many different permutations are there of 2 letters from "abcd"?

```python
import itertools
letters3 = 'abcd'

combinations3 = list(itertools.permutations(letters3, 2))

print(combinations3)
print(len(combinations3))
```

```python
#Total number of users

import json
 #read the JSON file
with open('profiles.json','r') as f:
    data = json.load(f)
```

```python
len(data)
print('Total number of users: ',len(data))
          
```

```python
# count the number of active users
active_user = 0
for user in data:
    if (user['isActive']) == True:
        active_user += 1    
print ('Total active users: ',active_user)        

    
```

```python
# Number of inactive users
in_active_user = 0
for user in data:
    if (user['isActive']) == False:
        in_active_user += 1    
print ('Total inactive users: ',in_active_user)
```

```python
def handle_commas(number_string):
    number_string = number_string.replace(',',"").replace('$',"")
    number = (number_string)
    return float(number)
```

```python
# Grand total of balances for all users
total_all_balances = 0
for user in data:
    total_all_balances += handle_commas(user['balance'])
print('Grand total of balances for all users: ',total_all_balances)   
```

```python
# Average balance per user
avg_balance_per_user = round((total_all_balances)/19,2)
print("Average balance per user: ",avg_balance_per_user)
```
